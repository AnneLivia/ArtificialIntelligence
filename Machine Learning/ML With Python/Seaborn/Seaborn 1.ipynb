```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Import data using builtin seaborn dataset

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
crash_df.info()
```

# Distribution plots

## Distribution Plot

A way to visualize univariate distribution. Provide distribution for one variable.
In mathematics, a univariate object is an expression, equation, function or polynomial involving only one variable. 

```python
# accidents where the person was not distracted
sns.distplot(crash_df['not_distracted'])
# that's a historgram plot
# the line is a kernel density estimation
```

```python
# accidents where the person was not distracted
sns.distplot(crash_df['not_distracted'], kde=False)
# that's a historgram plot
```

```python
# accidents where the person was not distracted
sns.distplot(crash_df['not_distracted'], bins=25)
# bins divide data between intervals.
```

# Joint plots

Used to compare two distributions and plots a scarrer plot by default

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde') 
# kde = kernel density estiamtion = infers the distribution of the data
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex') 
```

## KDE plots

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset

```python
sns.kdeplot(crash_df['alcohol'])
```

## Pair plots

Produces a matrix of relationships between each variable in your data for an instant examination of our data.
Plot the relationships across the entire dataframe in numerical values

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues') # in hue pass categorical data
```

## Rug plots

plot a single colluns of data points as sticks. we can se a more dense number of lines when data is more common. It is used to visualise the distribution of the data. 

```python
sns.rugplot(tips_df['tip'])
```

## Styling

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,6))
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')
plt.figure(figsize=(10,6))
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.set_context('talk', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8, 4))
sns.set_context('poster', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('white')
plt.figure(figsize=(10,8))
sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.despine(left=True, bottom=True) 
# removing axis, what happens to the bottom occurs to the top and what happens to left occurs to the right
```
