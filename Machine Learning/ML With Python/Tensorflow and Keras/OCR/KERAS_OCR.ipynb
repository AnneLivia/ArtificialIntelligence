```python
from keras_ocr import tools, pipeline
import matplotlib.pyplot as plt
```

keras-ocr will automatically download pretrained
weights for the detector and recognizer.

```python
pipeline = pipeline.Pipeline()
```

```python
speed_limit = tools.read('speed.jpg')
```

```python
plt.imshow(speed_limit);
```

```python
speed_limit.shape
```

```python
predictions = pipeline.recognize(images=[speed_limit])[0]
```

```python
display(predictions)
```

```python
# Plot the predictions
drawn = keras_ocr.tools.drawBoxes(
    image=speed_limit, boxes=predictions, boxes_format='predictions'
)
```

```python
plt.imshow(drawn)
```

```python
print(
    'Predicted:', [text for text, box in predictions]
)
```

```python
text = tools.read('text.jpg')
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(text);
plt.axis('off');
```

```python
recog2 = pipeline.recognize(images=[text])[0]
```

```python
# Plot the predictions
drawn2 = keras_ocr.tools.drawBoxes(
    image=text, boxes=recog2, boxes_format='predictions'
)
```

```python
plt.imshow(drawn2)
print(
    'Predicted:', [text for text, box in recog2]
)
```
