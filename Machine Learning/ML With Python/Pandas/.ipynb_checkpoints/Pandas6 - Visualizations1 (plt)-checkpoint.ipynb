```python
import matplotlib.pyplot as plt
# show plot inside notebook
%matplotlib inline 
import pandas as pd
import numpy as np
```

```python
# row a dice 5000 times. show the frquency and probability and histogram (total number of times it returns)

df_dice = pd.DataFrame(
    np.random.randint(1, 7, 5000),
    columns = ['Hist']
)

df_dice['Odds'] = df_dice['Hist'] + np.random.randint(1,7, 5000)

ax = df_dice.plot.hist(bins=12, alpha=0.88)
```

```python
ser = pd.Series(np.random.randn(1000), index=pd.date_range('11/15/2017', periods=1000))
ser = ser.cumsum()
ser.plot(alpha=0.88)
```

```python
dfdf = pd.DataFrame(np.random.randn(1000, 3), index=pd.date_range('11/15/2017', periods=1000), columns=list('ABC'))
dfdf = dfdf.cumsum()
dfdf.plot()
```

```python
pd.DataFrame(np.random.randn(5)).plot.bar()
```

```python
vals = ['A', 'B', 'C', 'D']
df = pd.DataFrame(np.random.rand(10, 4), columns=vals)
df.plot.bar()
```

```python
# Area plot 
# Define x range and y values
x_rng = range(1,15)
y_vals = [1,5,4,7,6,9,5,7,10,14,10,12,9,8]
# Change fill color and opacity
plt.fill_between(x_rng, y_vals, color="red", alpha=0.5)
plt.show()

```

```python
# Area plot with multiple areas
pd.DataFrame(np.random.rand(10,3), columns=['A','B','C']).plot.area()

```

```python
# Create a scatterplot with 100 random values
pd.DataFrame(np.random.rand(100,2), columns=['A','B']).plot.scatter(x='A', y='B')

```

```python
# Multiple column scatter plots
df_15 = pd.DataFrame(np.random.rand(50,4), columns=['A','B','C','D'])
ax = df_15.plot.scatter(x='A', y='B', color='DarkBlue', label='Grp 1')
df_15.plot.scatter(x='C', y='D', color='Orange', label='Grp 2', ax=ax)
```

```python
# Pie Charts with 4 random values
pd.Series(np.random.rand(4), index=['a','b','c','d'],name='Pie').plot.pie(figsize=(6,6))
```
