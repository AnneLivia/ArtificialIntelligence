```python
# reading from database

# first need to create a database connection
try:
    
    db_connection = pymysql.connect(db='mydb', user='root', passwd = '', host='localhost', port=3306)
    con = db_connection.cursor() # makes queries for the database
    sql = "INSERT INTO categoria (nome) VALUES ('suspense')"
    con.execute(sql)
    db_connection.commit()
    
    stud_df = pd.read_sql('SELECT * FROM categoria', con=db_connection)
    print(stud_df)
    
except Exception as e:
    
    print("Exception : {}".format(e))
    
finally:
    
    db_connection.close()
```

```python
# transform is going to execute a function in a dataframe
df_5 = pd.DataFrame({'A' : range(3), 'B' : range(1,4)})
df_5
```

```python
df_5.transform(lambda x : np.sqrt(x))
```

```python
df_5.transform(lambda x : np.exp(x))
```

```python
df_5.transform(lambda x : x ** 2)
```

```python
df_5.transform([lambda x : x ** 2, lambda x : x ** 3])
```

```python
df_5.transform({'A' : lambda x : x ** 2, 'B' : lambda x : x ** 3})
```

# group data

```python
dict_5 = {'Store' : [1, 2, 1, 2], 'Flavor' : ['Choc', 'Van', 'Straw', 'Choc'], 'Sales' : [26, 12, 18, 22]}
df_11 = pd.DataFrame(dict_5)
df_11
```

```python
# group rows a function can combine values
# getting the mean group by store
by_store = df_11.groupby('Store')
by_store.mean()

by_store.sum().loc[1]
```
