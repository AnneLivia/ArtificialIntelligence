# DATA VISUALIZATION 

## Checking Multicollinearity

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Checking correlation between independent variables
# Multicollinearity occurs when two or more independent variables are highly 
# correlated with one another in a regression model. 
# Multicollinearity can be a problem in a regression model because we would not 
# be able to distinguish between the individual effects of the independent 
# variables on the dependent variable.
plt.figure(figsize=(17, 13));
sns.heatmap(data=X.corr(), annot=True, cmap='PiYG_r');
plt.show()

# Energy and loudness = 0.77 correlation
```

```python
sns.scatterplot(y='energy', x='loudness', data=dataset);
# Positive correlation between energy and loudness. as loudness increases, so do energy.
```

```python
# Dropping correlated columns can fix the problem of multicollinearity
X.drop('energy', axis=1, inplace=True);
```

```python
fig, ax = plt.subplots(2, 7, figsize=(18, 4))
# tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area
fig.tight_layout();

i = 0;
j = 0;

# A scatter plot (aka scatter chart, scatter graph) uses dots to represent values 
# for two different numeric variables. The position of each dot on the horizontal 
# and vertical axis indicates values for an individual data point. Scatter plots 
# are used to observe relationships between variables.

for label, _ in dataset.iteritems():
    sns.scatterplot(y='song_popularity', x=label, data=dataset, ax=ax[i, j]);
    j+=1;
    if (j == 7):
        i = 1
        j = 0;

fig.show();
```

## SPLIT DATASET INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split;

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40);
```

# TRAINING MODELS

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor, ExtraTreesRegressor;
from sklearn.linear_model import LinearRegression;
from sklearn.neighbors import KNeighborsRegressor;
from sklearn.svm import SVR;
from sklearn.neural_network import MLPRegressor;
from xgboost import XGBRegressor;


# metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error;
```

### Linear Regression

```python
lrm = LinearRegression();
lrm.fit(X_train, y_train);
lrm_pred = lrm.predict(X_test);
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, lrm_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, lrm_pred)));
```

```python
# Normalizing for MLP
from sklearn.preprocessing import MinMaxScaler, StandardScaler;
from sklearn.pipeline import Pipeline;
```

### MLPRegressor

```python
model_neural = MLPRegressor(
    hidden_layer_sizes=(300, ),
    activation='relu',
    solver='adam',
    random_state=40,
    early_stopping=True
)
```

```python
mlp_std = Pipeline(steps=[
    ('stdscale', StandardScaler()),
    ('neural', model_neural)
])
```

```python
mlp_std.fit(X_train, y_train);

mlp_std_pred = mlp_std.predict(X_test);

print('===== Stardard Scale =====')
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, mlp_std_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, mlp_std_pred)));
```

```python
mlp_norm = Pipeline(steps=[
    ('norm', MinMaxScaler()),
    ('neural', model_neural)
])
```

```python
mlp_norm.fit(X_train, y_train);

mlp_norm_pred = mlp_norm.predict(X_test);

print('===== MinMax Normalization =====')
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, mlp_norm_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, mlp_norm_pred)));
```

### Decision Tree

```python
dtr = DecisionTreeRegressor(random_state=50, min_samples_leaf=400);
dtr.fit(X_train, y_train);
dtr_pred = dtr.predict(X_test);
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, dtr_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, dtr_pred)));
```

### AdaBoost Regressor

```python
ada = AdaBoostRegressor(random_state=40, n_estimators=100);
ada.fit(X_train, y_train);
ada_pred = ada.predict(X_test);
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, ada_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, ada_pred)));
```

### Random Forest Regressor

```python
rfr = RandomForestRegressor(random_state=40, n_estimators=500);
rfr.fit(X_train, y_train);
rfr_pred = rfr.predict(X_test);
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, rfr_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, rfr_pred)));
```

### Extra Trees Regressor

```python
ext = ExtraTreesRegressor(random_state=40, n_estimators=1000);
ext.fit(X_train, y_train);
ext_pred = ext.predict(X_test);
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, ext_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, ext_pred)));
```

### XGBoost Regressor

```python
xgb = XGBRegressor();
xgb.fit(X_train, y_train);
xgb_pred = xgb.predict(X_test);
print('Mean Squared Error: {}'.format(mean_squared_error(y_test, xgb_pred)));
print('Mean Absolute Error: {}'.format(mean_absolute_error(y_test, xgb_pred)));
```
